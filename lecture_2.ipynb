```python
from google.colab import files
import pandas as pd
import io  # Import the io module

# Upload the CSV file
uploaded = files.upload()

# Get the filename
file_name = list(uploaded.keys())[0]

# Read the CSV file into a DataFrame
df = pd.read_csv(io.StringIO(uploaded[file_name].decode('utf-8')))

# Display the DataFrame
df
```

```python
%matplotlib inline
# now here we are going to add the x and the y labels for  the given plot that we have made
plt.xlabel('area(Sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
# now from the above plot we can say that the plot and the data are suitable for the linear regression module plotting and hence firstl we will be creating a object of the linear regression
# module
# remember that fitting the data meaning we are training the given linear regression model based on the available data sets and the available data points
# and the first argument that we have to pass in that is the 2D array thta is the area and the second argument which iss our y axis that iss price
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
# here we receive no error that means out linear regression model is ready to make the predictions and hence we can now predict the prices of the houses
```

```python
import numpy as np

# Reshape the input to a 2D array
predicted_price = reg.predict(np.array([[3300]]))

# Display the predicted price
print(predicted_price)
# so for any linear equation there is a slope/coefficient and an intercept. and hence the model is basically calcculating the value for the m and b in the equation y=mx+c
# hence if i do reg_coef that should show me the value of the slope or the coefficient or m  nd the same iss  true for the intercepts
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.xlabel('area', fontsize=20)
plt.ylabel('prices', fontsize=20)

# Scatter plot of data points
plt.scatter(df.area, df.prices, color='red', marker='+')

# Plot the regression line
plt.plot(df.area, reg.predict(df[['area']]), color='blue')

plt.show()

```

#Linear Regression with the multiple varibales

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
from google.colab import files
import pandas as pd
import io  # Import the io module

# Upload the CSV file
uploaded = files.upload()

# Get the filename
file_name = list(uploaded.keys())[0]

# Read the CSV file into a DataFrame
df = pd.read_csv(io.StringIO(uploaded[file_name].decode('utf-8')))

# Display the DataFrame
df
# as here we have one nan point and we haave to handle it hence what we will do is we wwill simplr consider the mediam of the rest of the values from the same column and then put it
# into the place of the nan value
```

```python
# Display the column names in the DataFrame
print(df.columns)
```

```python
# Assuming 'bedrooms' is the correct column name
median_bedroom = df['bedroom'].median()
print("Median number of bedrooms:", median_bedroom)
# now i want to get the value of the median in the whole number and hence i will import the math module and perform the operation
```

```python
import math
median_bedroom=math.floor(df.bedroom.median())
print("Median number of bedrooms:", median_bedroom)
```

```python
df.bedroom.fillna(median_bedroom)
# here we have simply upated the value of the na to the 3 and now we will now insert this value to the main table of the data
```

```python
df.bedroom=df.bedroom.fillna(median_bedroom)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedroom','age']],df.price)
# df.price is my target variable
# here our model is ready and now we will do our predictions on the model
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
# here as the house is 40 yrs old and hence the price of the house is also less and compaired to the other houses
```
